# ***PHÂN SỐ***

---

Cho 2 phân số đúng và tối giản image.png . Mỗi phép biến đổi là tăng **a** và **b** lên 1, sau đó giản ước phân số nhận được.

Hãy xác định sau bao nhiêu bước biến đổi từ phân số thứ nhất ban đầu nhận được phân số thứ 2 đã cho.

**Dữ liệu:** Vào từ thiết bị nhập chuẩn gồm 4 dòng, mỗi dòng chứa 1 số nguyên **a, b, c, d**, 0 < **a** < **b** ≤10^5, 0 < **c** < **d** ≤ 10^5, **a** và **b** nguyên tố cùng nhau, **c** và **d** nguyên tố cùng nhau image.png 

**Kết quả:** Đưa ra thiết bị xuất chuẩn số 0 nếu không có cách biến đổi hoặc một số nguyên – số lượng phép biến đổi.

## **Abstract**

---

> Cho 2 phân số ***a/b*** và ***c/d***, **tìm số lần biến đổi** phân số **a/b** để thu được phân số ***c/d***.
 Biết mỗi lần biến đổi sẽ tăng tử và mẫu phân số ***a/b*** thêm 1 sau đó rút gọn phân số.

## **Pattern recognization**


---
***BRUTE FORCE***
---

Với giới hạn cho các số mà đề đưa ra không quá lớn, mọi giá trị đều nằm trong khoảng < 10^5 (***a, b, c, d***, 0 < ***a*** < ***b*** ≤10^5, 0 < ***c*** < ***d*** ≤ 10^5, ***a*** và ***b*** nguyên tố cùng nhau, ***c*** và ***d*** nguyên tố cùng nhau)

Ta dùng phương pháp brute force (vét cạn) để tìm đáp án, nếu không có đáp án thì xuất 0.

>**Ý tưởng**: 
>>Dùng vòng lặp while để thực thi việc biến đổi phân số ***a/b*** lần lượt đến khi ***a/b*** biến đổi thành phân số ***c/d*** thì xuất ra số bước biến đổi.
>
>>Nếu không tìm ra kết quả thì xuất ra tương ứng số bước là 0.
---
Ta dùng 1 ví dụ, giả sử ta có ***a/b*** = 1/6 và ***c/d*** = 2/3.
>Thực hiện lần lượt từng bước biến đổi cho đến khi tìm được kết quả bài toán như sau:
>
>>vòng lặp 1: a/b = 1(+1)/6(+1) = 2/7 . Không rút gọn được. Chưa bằng số c/d.
>
>>vòng lặp 2: a/b = 2(+1)/7(+1) = 3/8 . Không rút gọn được. Chưa bằng số c/d.
>
>>vòng lặp 3: a/b = 3(+1)/8(+1) = 4/9 . Không rút gọn được. Chưa bằng số c/d.
>
>>vòng lặp 4: a/b = 4(+1)/9(+1) = 5/10. Rút gọn thành a/b = 1/2. Chưa bằng số c/d.
>
>>vòng lặp 5: a/b = 1(+1)/2(+1) = 2/3 . Bằng phân số c/d. 
>
> Kết thúc vòng lặp và xuất kết quả.










## **Algorithm designed**


---

**Initialize:**
>***n*** = 0 - là số nguyên tố, dùng để tính số bước lặp lại của việc biến đổi.

**Input:**
>***a, b, c, d*** là số nguyên tố (***a, b, c, d***, 0 < ***a*** < ***b*** ≤10^5, 0 < ***c*** < ***d*** ≤ 10^5).

**Output:**
>***n*** - số bước để biến đổi ***a/b*** thành ***c/d***.

**Bắt đầu**

>>**Bước 1:** Ta lập 1 hàm để tìm giá trị ước chung lớn nhất có đầu vào là ***a*** và ***b***:
```
Dùng thuật toán Euclid để tìm UCLN:
   if (b == 0):
      return a
   return ucln(b, a % b) 
```
>
>>**Bước 2:** Kiểm tra xem phân số ***a/b*** có lớn hơn phân số ***c/d*** hay không. Nếu lớn hơn thì không vào vòng lặp while mà xuất ra luốn kết quả ***n***. Nếu không thì tới Bước 3.
>
>>**Bước 3:** Ta thực hiện vòng lặp while để tìm ***n***:
```
    1. biến đổi a/b:
      1.1/ Tăng tử và mẫu của phân số a/b lên 1:
        a += 1
        b += 1
      1.2/ Rút gọn phân số a/b bằng cách chia tử và mẫu cho UCLN
        a /= UCLN(a,b)
        b /= UCLN(a,b)
    2. Sau khi biến đổi xong ta tăng n lên 1 đơn vị:
        n += 1
    3. Kiểm tra phân số a/b có bằng c/d chưa, nếu rồi thì xuất n
    4. Lặp lại
```
>
>>**Bước 4:** Xuất kết quả: 
```
n
```

**Kết thúc.**

**Time Complexity: O(nlog(n))**

Dùng thuật toán Euclid để tìm UCLN: ***O(log(n))***

Vòng lặp while : ***O(n)***

Vì bước dùng thuật toán Euclid có nằm trong vòng lặp while nên độ phức tạp thuật toán sẽ là: O(n) x O(log(n)) = ***O(nlog(n))***


## **CODE**

---



````
a = int(input())
b = int(input())
c = int(input())
d = int(input())
def ucln(a, b):
    if (b == 0):
        return a
    return ucln(b, a % b)

if (a*d < b*c and c/d < 1 and a/b < 1):
  n = 0
  while (a*d <= b*c):
      a = a+1
      b = b+1
      uoc = ucln(a,b)
      a = (a)//uoc
      b = (b)//uoc
      n+=1
      if (a==c and b == d):
        print(n)
        quit()
print(0)
````
