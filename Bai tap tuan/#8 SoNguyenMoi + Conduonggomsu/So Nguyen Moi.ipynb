# **SỐ NGUYÊN MỚI**

---

Cho số nguyên dương **n** có không quá 100 chữ số. Hãy xác định số nguyên lớn nhất **m** chia hết cho 3 và khác **n** ở đúng một chữ số.

Ví dụ: **n** = 123 thì **m** sẽ là 723.

**Dữ liệu:** Vào từ thiết bị nhập chuẩn gồm một dòng chứa số nguyên **n** có không quá 100 chữ số và không chứa các số 0 không có nghĩa.

**Kết quả:** Đưa ra thiết bị xuất chuẩn số nguyên **m** tìm được.

## **Abstract**

---

Cho số nguyên **n** (0 ≤ **n** ≤ 10^100), tìm số nguyên lớn nhất **m** chia hết cho 3 và khác **n** ở đúng 1 chữ số. 

***Ví dụ:*** 

>Cho n = 123. 
>>Số nguyên m lớn nhất chia hết cho 3 và khác n ở 1 chữ số là m = 723.
>
>Cho n = 999.
>>Số nguyên m lớn nhất chia hết cho 3 và khác n ở 1 chữ số là m = 996.

##**Pattern recognization**


---
**Dynamic Programming (Quy hoạch động)**
---

Dựa vào đề bài toán, ta có thể dễ dàng nhận ra những bài toán con ta cần xử lý là:

**Bài toán con 1:**
>Dựa vào **n**, tìm số nguyên **m** lớn nhất chia hết cho 3.

**Bài toán con 2:**
>Từ số nguyên **m** đã tìm được, thay đổi 1 chữ số của **m** để nó vẫn giữ nguyên tính chất đề bài là: lớn nhất chia hết cho 3, và đồng thời khác **n** 1 chữ số.

Vì bài toán thể hiện được tính chất của **các bài toán con gối nhau** và **cấu trúc con tối ưu** nên ta sẽ sử dụng **Dynamic Programming** để giải bài toán.

**Ý tưởng:**

**Bài toán con 1**: 
>Tìm số nguyên **m** lớn nhất chia hết cho 3. Ta dựa vào việc kết hợp các tính chất toán học: Phép chia lấy dư và Phép cộng. Đồng thời xử lý từng chữ số trong **n** theo chiều từ phải sang trái để tìm **m** thỏa yêu cầu đầu tiên của đề bài.
>>Ví dụ: Cho số **n** = 123. 
>
>>Ta xét chữ số ở vị trí đầu tiên: **n**[0] = 1. Vậy để cho số nguyên **m** tạo ra là lớn nhất và chia hết cho 3, ta sẽ cộng thêm vào **n**[0] 1 con số mà **n** sau khi cộng thì vẫn thỏa yêu cầu của bài toán con 1 và riêng chữ số sau khi cộng không được quá 10, ở đây ta cộng vào **n**[0] là 6. 
>
>>Do đó m thu được là **m** = 723. Số lớn nhất thu được từ **n** và chia hết cho 3.

**Bài toán con 2**: 
>Tìm số nguyên **m** thỏa yêu cầu đề và khác với **n** 1 chữ số. Ở đây tốt nhất ta chỉ cần kiểm tra các trường hợp mà các số trong **n** đều là số lớn nhất và chia hết cho 3, từ đó nó sẽ không xét qua bài toán con 1 mà chuyển hẳn sang bài toán con 2.
>>Ví dụ: Lấy tiếp ví dụ ở trên, **m** thu được là **m** = 723.
>
>>Nếu xét với số **n** ban đầu thì chắc chắn là nó có thay đổi ngay số được cộng. Do đó ta chỉ cần xuất ra **m**.







## **Algorithm designed**


---
**Input:**
>***n*** Số nguyên

**Output:**
>**m**: Số nguyên

**Initialize:**
>
```
a = list(n) : Tách các chữ số trong số nguyên n ra và lưu vào mảng a

sodu = int(n)%3 : Lưu số dư của n sau khi chia cho 3

check = 1 : Biến dùng để dừng vòng lặp

td = 0 : Biến dùng để kiểm tra cho bài toán con 2

dic = {2:[7,4,1], 1:[8,5,2], 0:[9,6,3]} : Dictionary được tạo ra dựa vào số dư sau khi chia n cho 3, ta có thể tìm ra các giá trị dùng để cộng thêm vào các chữ số trong khi xét ở bước kiểm tra cho bài toán con 1, ta để ý là tương ứng với các số được cộng vào thì chữ số sau khi cộng luôn luôn chia hết cho 3. 

VD: Ta có n = 992. Có thể dễ dàng nhận thấy là n khi chia lấy dư cho 3 thì còn 
dư lại là 2. Lúc đó đối chiếu vào dictionary, ta vào dic[2], lấy ra được list 3 số tương ứng [7,4,1]. Sau đó xét từng chữ số 1 theo chiều từ trái sang phải để tìm kết quả cho bài toán con 1 như sau:
- Xét chữ số đầu tiên là 9 -> 9 + 7/4/1 đều lớn hơn 9, ta xét qua chữ số tiếp theo.
- Xét 9 -> 9 + 7/4/1 đều lớn hơn 9, ta xét qua chữ số tiếp theo.
- Xét 2 -> 2 +7 = 9, không lớn hơn 9. Đã có sự thay đổi nên ta không lặp nữa.
=> Số sau khi thay đổi là 999.

```

**Bắt đầu**

**Bước 1:**

>Ta thực hiện bài toán con thứ 1 ở bước này, bắt đầu là chạy vòng lặp for để duyệt mảng a. Mục đích là xét từng chữ số trong n, có số nào có thể thay đổi để làm cho số nguyên m tạo ra từ n  thỏa được yêu cầu trong bài toán con 1 hay không. Nếu có thì ngừng lặp.

```
      a[i] = int(a[i])
      #Dựa vào số dư của n sau khi chia cho 3, dùng vòng lặp để gọi đến từng phần tử trong mảng ứng với cơ số là số dư đó. 
      for j in dic[sodu]: 
         #Kiểm tra xem a[i] sau khi cộng với các phần tử trong mảng đó,thì nó có < 10 hay không, nếu có thì thực hiện các lệnh thay đổi cả biến a[i], td, và check. 
          Mục đích kiểm tra là các số sau khi cộng vẫn < 10 thì sau khi thay đổi, n sẽ chia hết cho 3 và đồng thời nó là số lớn nhất tìm được.
          if (a[i]+j) < 10:
             a[i] = a[i]+j
             td = 1
             check = 0
             break
      #Nếu có sự thay đổi, thoát vòng lặp.
      if check == 0:
             break
```


**Bước 2:**

>Sau khi xét xong bài toán con 1, ta đi tới kiểm tra bài toán con 2. Nếu sau bước 1 mà n không thay đổi, thì ta trừ đi chữ số ở vị trí cuối của n theo 2 trường hợp.

```
if td==0:
        if sodu!=0:
            a[-1] = int(a[-1])-sodu
        else:
            a[-1] = int(a[-1])-3
```

**Bước 3:** 
>Chuyển mảng a sau khi xử lý thành dạng số nguyên, gán vào **m** và xuất **m**.

**Kết thúc.**

**Time Complexity: O(n)**

Không kể các lệnh đơn có độ phức tạp là O(1) thì trong code có dứa 2 vòng lặp for lồng nhau tại bước 2, lần lượt có độ phức tạp là O(n) và O(1).

Cho nên ta sẽ có độ phức tạp cho bài toán như sau: O(n) * O(1) = **O(n)**


## **CODE**

---



````
n = input()
a = list(n)
sodu = int(n)%3
check = 1
td = 0
def strtoint(integers):
      strings = [str(integer) for integer in integers]
      a_string = "".join(strings)
      an_integer = int(a_string)
      return an_integer

dic = {2:[7,4,1], 1:[8,5,2], 0:[9,6,3]}

for i in range (0,len(a)):
      a[i] = int(a[i])
      for j in dic[sodu]:
          if (a[i]+j) < 10:
             a[i] = a[i]+j
             td = 1
             check = 0
             break
      if check == 0:
             break
if td==0:
        if sodu!=0:
            a[-1] = int(a[-1])-sodu
        else:
            a[-1] = int(a[-1])-3
m = strtoint(a)
print(m)
````
