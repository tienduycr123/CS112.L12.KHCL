## ***ĐÈN LED***

Alice vừa hoàn thành một hợp đồng lắp bảng hiển thị số bằng đèn LED. Mỗi chữ số được hiển thị trong một khung chữ nhật với 7 ống đèn LED. Bằng cách bật các ống đèn LED thích hợp ta có thể hiển thị chữ số bất kỳ. Số đèn sáng càng nhiều thì việc hiển thị chữ số đó càng tốn năng lượng. Ví dụ, hiển thị chữ số 9 sẽ tốn năng lượng hơn hiển thị chữ số 7.

image.png

Sau khi bàn giao sản phẩm trong tay Alice còn thừa lại một số khá nhiều các khung hiển thị số và một cục pin nguồn. Dung lượng pin cho phép bật sáng **n** ống đèn LED. Alice muốn dùng pin bật đúng **n** ống đèn để hiển thị một số và số hiển thị được phải có tổng chữ số là lớn nhất.

Hãy xác định tổng lớn nhất của các chữ số của số có thể bật sáng.


**Dữ liệu:** Vào từ thiết bị nhập chuẩn gồm một dòng chứa số nguyên **n** ($2$ ≤ **n** ≤ $10^6$).

**Kết quả:** Đưa ra thiết bị xuất chuẩn một số nguyên – tổng lớn nhất đạt được..

#**DECOMPOSITION**

Như đề đã cho như sau
>"Sau khi bàn giao sản phẩm trong tay Alice còn thừa lại một số khá nhiều các khung hiển thị số và một cục pin nguồn. Dung lượng pin cho phép bật sáng **n** ống đèn LED. Alice muốn dùng pin bật đúng **n** ống đèn để hiển thị một số và số hiển thị được phải có tổng chữ số là lớn nhất."

Thì bài toán này sẽ dựa vào số **n** ống đèn mà Alice được cho.

Vì vậy ta sẽ bắt đầu phân tích sự liên quan giữa số ống đèn và yêu cầu bài toán là số hiển thị phải có tổng chữ số lớn nhất.

Đầu tiên, ta tổng hợp số ống đèn mà mỗi số hiển thị sẽ sử dụng dựa vào hình sau:

image.png

Như hình trên, ta sẽ có được số đèn mà từng chữ số lúc hiển thị sẽ được sử dụng:

> Số: 0 - Số đèn sử dụng là m = 6

> Số: 1 - Số đèn sử dụng là m = 2

> Số: 2 - Số đèn sử dụng là m = 5

> Số: 3 - Số đèn sử dụng là m = 5

> Số: 4 - Số đèn sử dụng là m = 4

> Số: 5 - Số đèn sử dụng là m = 5

> Số: 6 - Số đèn sử dụng là m = 6

> Số: 7 - Số đèn sử dụng là m = 3

> Số: 8 - Số đèn sử dụng là m = 7

> Số: 9 - Số đèn sử dụng là m = 6

Sau khi đã tổng hợp, ta sắp xếp lại các thứ tự các chữ số qua số **m** ống đèn.

> m = 2, có thể mở được đèn hiện số: 1

> m = 3, có thể mở được đèn hiện số: 7

> m = 4, có thể mở được đèn hiện số: 4

> m = 5, có thể mở được đèn hiện số: 5, 3, 2

> m = 6, có thể mở được đèn hiện số: 9, 6, 0

> m = 7, có thể mở được đèn hiện số: 8

Vì Alice cần phải tối ưu tổng số thu được với **n** ống đèn, cùng với list tổng hợp ở trên. Ta có thể chọn ra số **m** sẽ thu được tổng lớn nhất nếu sử dụng. Ở đây ta sẽ chọn **m** = 3, vì với **m** = 3, ta sẽ thu được tổng số lớn nhất từ **n** đã cho. Ta cứ hiểu nôm na là từ **n** ống đèn đã cho, ta lấy ra **m** ống đèn để tạo thành chữ số 7, cứ bao nhiêu ống đèn thì sẽ có tổng n/m lần số ống đèn nhân cho 7.

Giả sử ta cho số **n** = 6, tương ứng với nếu mở đèn với số đèn hiển thị là lớn nhất, đối chiếu vào trên list tổng hợp thì ta sẽ được số 9. Nhưng với việc lấy tổng lớn nhất, thì ta phải dùng phép **chia lấy nguyên**: 

> n = 6, m = 3, số thu được qua m = 3 là 7. Ta sẽ có phép tính n/m = 6/3 = 2. 2*7 = 14. 

> Tổng số lớn nhất thu được sẽ là 14 nếu ta cứ sử dụng 3 ống đèn lần lượt để tạo thành các số 7.

Lúc này ta sẽ có 2 trường hợp:

**DƯ 1** tức là ống đèn còn lại chỉ có duy nhất 1 ống đèn, không tạo ra được chữ số nào cả. Nhưng chúng ta vẫn không thể bỏ phí ống đèn đó.
>>*Giải pháp*: Ta sẽ rút đi 3 ống trong n ống ngay từ đầu, rồi vẫn tính toán như bình thường, nhưng cuối cùng tổng số sẽ cộng cho 4 (Vì với 4 ống đèn sẽ tạo dược chữ số 4), xét theo toán học thì là trừ đi 1 cho kết quả của phép tính n/m với m = 3 và cộng cho 4 vào tổng Sum1.
>
> VD: n = 7, nếu theo phép tính n/3 = 7/3 = 2 dư 1, tổng thu được sẽ là Sum = 2*7 = 14. Nhưng theo giải pháp ta sẽ có 1 cách khác:
>> Rút 3 từ n, n lúc này sẽ còn 4, thực hiện phép tính: 4/3 = 1 dư 1, tổng thu được sẽ là Sum = 1*7 + 4 = 11. 11 sẽ là kết quả cho trường hợp này.

**DƯ 2**, tức là ống đèn còn lại 2 ống đèn. 
>>*Giải pháp*: Lúc này thì ta chỉ cần cộng thêm 1 vào tổng số sau khi tính toán (Vì với 2 ống đèn dư kia sẽ tạo được thành chữ số 1).
>
>VD: n = 5, ta sẽ có tổng là Sum = 7(n/3) + 1 = 7*(5/3) + 1 = 8.

#**ALGORITHM DESIGN**
Input: 
> n - số nguyên


Output:
> tong 

Initialize:
```
x = int(n/3) - Lưu trữ kết quả cho phép chia lấy nguyên n/3 
y = n%3 - Lưu trữ kết quả cho phép chia lấy dư n%3
tong  = 0 - Chứa kết quả của Output
```
**Thuật toán**

Mã giả:
```
if (y là 2) then
  tong := x*7 + 1
else if (y là 1) then
  tong := (x-1)*7 + 4
else if (y là 0) then
  tong := x*7
```




Vì code mà nhóm đã xây dựng không có vòng lặp nào nên độ phức tạp thuật toán tổng lại sẽ là

**Time Complexity: O(1)**

#**CODE**


````
n = int(input())
x = int(n/3)
y = n%3
tong = 0
if (y == 2):
  tong = x*7 + 1
elif (y == 1):
  tong = (x-1)*7 + 4
elif (y == 0):
  tong = x*7
print(tong)
````
