## ***TẢO BIỂN(SEAWEED)***

*Tảo biển sinh sản rất nhanh khi có môi trường thuận lợi với chúng và có những loài còn tiết ra môi trường những chất độc hại.*

Một loại tảo nâu trong môi trường nước bị ô nhiễm nặng sinh sản theo quy luật sau:

* Ngày đầu tiên (ngày 0) có n cá thể ở mức 1,
* Ở mỗi ngày tiếp theo, mỗi cá thể mức i sinh ra i cá thể mức 1, các cá thể mới sinh sẽ sinh sôi, phát triển từ ngày hôm sau.
Bản thân các cá thể mức i phát triển thành mức i+1 và chu kỳ phát triển trong ngày chấm dứt.
* Hãy xác định sau k ngày trong nước biển có bao nhiêu cá thể.





**Dữ liệu:** Vào từ thiết bị nhập chuẩn gồm một dòng chứa 2 số nguyên n và k (1 ≤ n ≤ 1000, 1 ≤ k ≤ 10^18).

**Kết quả:** Đưa ra thiết bị xuất chuẩn một số nguyên – số lượng cá thể  tảo theo mô đun 10^9+7.

**ABSTRACT**

Given $n$ elements are both equal to 1 and $k$ iterations.
With condition: in each loop, $ a_i $ generates $ a_i $ price elements with value is 1, and the concurrent value of $ a_i $ has been increased by 1. Find the total number of elements of the sequence after $ k $ iterations.










**DECOMPOSITION**

After $ k $ iterations:
>Loop 1: $ 1_ {1} = 1 $

>Loop 2: $ 1_ {1} + 1_ {2} = 2 $

>Loop 3: $ 3_ {1} + 1_ {2} + 1_ {3} = 5 $

>Loop 4: $ 8_ {1} + 3_ {2} + 1_ {3} + 1_ {4} = 13 $ 

> ...

If we compare the sequence that we got from above: $ 1, 2, 5, 13, ... $, 
We can easily see that those numbers all match the numbers in the fibonacci sequence starting with 1: $ 1, 1, 2, 3, 5, 8, 13, ... $

Based on that, we can understand that the problem we are looking for is to find the value of the position $(2 * k)$ in the ***fibonacci sequence*** starting with $1$. Then multiply by $n$.

**CODE**
>$Solution1$: Using *Space Optimized Fibonacci*.

````
def taobien(x, k):
  n1=0
  n2=x
  n3=0
  sum1=0
  for i in range(0,k*2+1):
      sum1=n1+n2
      n1=n2
      n2=sum1
      n3=n3+1
      if (i==k*2):
        return n1
        break
n, k = map(int, input().split())
kq = taobien(n,k)
m = 1000000007
print(kq%m)
````

>$Solution2$: Using Matrix to solve the problem. 

````
F = [[1, 1], [1, 0]]
module = int(1e9+7)
def nhanmatran(a, b, m): 
    a00 = (((a[0][0] * b[0][0]) % m) + ((a[0][1] * b[1][0]) % m)) % m 
    a01 = (((a[0][0] * b[0][1]) % m) + ((a[0][1] * b[1][1]) % m)) % m
    a10 = (((a[1][0] * b[0][0]) % m) + ((a[1][1] * b[1][0]) % m)) % m
    a11 = (((a[1][0] * b[0][1]) % m) + ((a[1][1] * b[1][1]) % m)) % m
    a[0][0] = a00 
    a[0][1] = a01 
    a[1][0] = a10 
    a[1][1] = a11 
def luythua(a, k):
    if k == 0 or k == 1: 
        return; 
    FC = [[1, 1], [1, 0]]; 
    luythua(a, k // 2) 
    nhanmatran(a, a, module)     
    if (k % 2 != 0): 
        nhanmatran(a, FC, module) 

n, k = map(int, input().split())
#Tim gia tri cua day fibonacci o vi tri 2*k
luythua(F, 2*k) 
kq = (n*F[0][0]) % module
print(kq)  
````
