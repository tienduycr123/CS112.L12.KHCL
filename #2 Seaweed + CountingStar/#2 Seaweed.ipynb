## ***TẢO BIỂN(SEAWEED)***

*Tảo biển sinh sản rất nhanh khi có môi trường thuận lợi với chúng và có những loài còn tiết ra môi trường những chất độc hại.*

Một loại tảo nâu trong môi trường nước bị ô nhiễm nặng sinh sản theo quy luật sau:

* Ngày đầu tiên (ngày 0) có n cá thể ở mức 1,
* Ở mỗi ngày tiếp theo, mỗi cá thể mức i sinh ra i cá thể mức 1, các cá thể mới sinh sẽ sinh sôi, phát triển từ ngày hôm sau.
Bản thân các cá thể mức i phát triển thành mức i+1 và chu kỳ phát triển trong ngày chấm dứt.
* Hãy xác định sau k ngày trong nước biển có bao nhiêu cá thể.





**Dữ liệu:** Vào từ thiết bị nhập chuẩn gồm một dòng chứa 2 số nguyên n và k (1 ≤ n ≤ 1000, 1 ≤ k ≤ 10^18).

**Kết quả:** Đưa ra thiết bị xuất chuẩn một số nguyên – số lượng cá thể  tảo theo mô đun 10^9+7.

**Abstract**



> Given the Fibonacci array starting at **n**, find the value at **2*k** position in that array.



**Pattern recognization**

> Space Optimized Fibonacci





**Algorithm designed**

Input: **n**, **k** is an integer, *1 <= **n** <= 1000, 1 <= **k** <= 10^18*

*a* is an fibonacci array starting at **n**, *a* can be represented as *a0*, *a1*, *a2*,... *ai* (with i = **2*k + 1**)

Output: value in **a[2*k]** of that array with 1000000007 module

Initialize:
```
  n1 = 0
  n2 = n    #To make n is the first value in the array 
  n3 = 0
  sum1 = 0
```
Make a Def: Loop *i* from 0 to **2*k+1** for each element of the array *a*:
```
      #From n, make a fibonacci array:  
      sum1 = n1+n2
      n1 = n2
      n2 = sum1
      n3 = n3+1 

      When i reachs to 2*k return the value in a[2*k]
```

Print:

```
(Value in **a[2*k]** which has been found)%1000000007
```
End.

**Time Complexity: O(2*k)**

**CODE**

````
def taobien(x, k):
  n1=0
  n2=x
  n3=0
  sum1=0
  for i in range(0,k*2+1):
      sum1=n1+n2
      n1=n2
      n2=sum1
      n3=n3+1
      if (i==k*2):
        return n1
        break
n, k = map(int, input().split())
kq = taobien(n,k)
m = 1000000007
print(kq%m)
````
